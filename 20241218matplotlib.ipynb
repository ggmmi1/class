```python
font1 = {'family': 'fantasy', 'color': 'deeppink', 'weight': 'normal', 'size': 'xx-large'}
plt.plot([1,2,3,4], [1,4,9,16])
plt.xlabel('x', labelpad=30, loc='left', fontdict=font1)
plt.ylabel('y', loc='top')
plt.show()
```

```python
#범례 표시하기
line1, = plt.plot([1, 2, 3], label='label1')
line2, = plt.plot([2, 4, 6], label='label2')
line3, = plt.plot([3, 6, 9], label='label3')
plt.legend(handles=[line1, line2, line3], loc=(0,1)) #loc가 범례표시하는 것 loc(1,1) 오른족 위 loc(0.5, 0.5)가운데 등
plt.show()
```

```python
#범례 표시학기 2
line1, = plt.plot([1, 2, 3], label='label1')
line2, = plt.plot([2, 4, 6], label='label2')
line3, = plt.plot([3, 6, 9], label='label3')
plt.legend(ncol=3, frameon=True) #ncol 범례 열 설정 #frameon 테두리 설덩 True있, False없
plt.show()
```

```python
#타이틀 설정하기
line1, = plt.plot([1, 2, 3], label='label1')
line2, = plt.plot([2, 4, 6], label='label2')
line3, = plt.plot([3, 6, 9], label='label3')
plt.legend(ncol=3, frameon=True)
plt.title("legand test")
plt.grid(True,axis="both")
plt.show()
```

```python
#막대 그래프 그리기 1

import numpy as np
x = np.arange(3)

plt.bar(x,[100, 400, 900], color=["r", "g", "b"], width=0.6)
plt.plot(x)

plt.xticks(x, ['2018', '2019', '2020']) #눈금 표시
plt.show()
```

```python
#막대 그래프 그리기 1-2

import numpy as np
x = np.arange(3)
y = [100, 400, 900]

plt.bar(x,y, color=["r", "g", "b"], width=0.6)
plt.plot(x,y)

plt.xticks(x, ['2018', '2019', '2020']) #눈금 표시
plt.show()
```

```python
#막대 그래프 그리기 2

y = np.arange(3)
x = [100, 400, 900]

plt.barh(y,x,color=["r", "g", "b"], height=0.7)


plt.yticks(y, ['2018', '2019', '2020']) #눈금 표시
plt.show()
```

```python
#막대 그래프 그리기 2-2

y = np.arange(3)
x = [100, 400, 900]

plt.barh(y,x,color=["r", "g", "b"], height=0.7)
plt.plot(x,y)


plt.yticks(y, ['2018', '2019', '2020']) #눈금 표시
plt.show()
```

```python
#산점도 그리기
n = 50 #50개를 그리겠다.
x = np.random.rand(n)
y = np.random.rand(n)

area = (30 *np.random.rand(n))**2 #0과 30의 난수
colors = np.random.rand(n)

plt.scatter(x, y, s=area, c=colors, cmap="Spectral", alpha=0.5) #alpha 투명도
plt.show()
```

```python
#산점도 그리기 2
x = [2,4,2]
y = [0,3,4]

area = [1000,2000,3000]

plt.scatter(x, y, area, alpha=0.5) #alpha 투명도
plt.show()
```

```python
#산점도 그리기 2
x = [1,4,6]
y = [2,3,4]

area = [1000,2000,3000]
plt.plot(x,y)

plt.scatter(x, y, area, alpha=0.5) #alpha 투명도
plt.show()
```

```python
#파이차트 그리기
ratio = [34, 32, 16, 18]
labels = ['Apple', 'Banana', 'Melon', 'Grapes']
wedgeprops = {'width':0.7, 'edgecolor': 'k', 'linewidth': 1}

plt.pie(ratio, labels=labels, autopct='%.1f%%', explode=[0, 0.1, 0, 0.1], colors=['red', 'yellow', 'green', 'purple'], wedgeprops=wedgeprops)
plt.show()
```

```python
#실습1
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
sales_2019, = plt.plot(months,[100,120,140,110,130,150,160,170,180,200,190,210], label='2019')
sales_2020, = plt.plot(months,[90,110,130,120,140,160,170,160,150,180,200,190], label='2020')

plt.legend(handles=[sales_2019, sales_2020],ncol=1)
plt.xlabel("month")
plt.ylabel("center")
plt.title("Monthly Sales Comparison (2019-2020)")
plt.show()
```

```python
#실습2
import numpy as np
x = np.arange(5)

categories = ['Category 1','Category 2','Category 3','Category 4','Category 5']
data = [20, 35, 15, 27, 45]

plt.bar(categories, data)
plt.title("Bar Chart")
plt.ylabel("Values")
plt.xlabel("Categories")
plt.grid()
plt.xticks(x, categories, rotation = 45)
plt.yticks([0,10,20,30,40,50]) #plt.ylim(0,50)으로 대체 가능
plt.show()
```

```python
#실습 3
labels = ['Apple', 'Banana', 'Mango', 'Blueberry']
sizes = [15, 30, 45, 10]
wedgeprops = {'width':0.7, 'edgecolor': 'k', 'linewidth': 1}


plt.pie(sizes, labels=labels, autopct='%.1f%%', explode=[0, 0.1, 0, 0], colors=['blue', 'orange', 'green', 'red'], startangle=90, shadow=True)
plt.show()
```

```python
#실습 4

x = np.linspace(0, 10, 100) #0에서 10을 100개로 나눔

y_sin = np.sin(x)
y_cos = np.cos(x)

sin_line, = plt.plot(x, y_sin, color='blue', label='sin(x)')
cos_line, = plt.plot(x, y_cos, color='red', label='cos(x)')

plt.legend()
plt.grid(True,axis='both')
plt.xlabel("x")
plt.ylabel("y")
plt.title("Sine and Cosine Functions")
plt.show()
```
