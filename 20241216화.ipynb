```python
#DataFrame 방법1
index = ["2018", "2019", "2020", "2021"]
data = {
    "영희" : [143, 150, 157, 160],
    "철수" : [165, 172, 175, 180]
}
growth = pd.DataFrame(data, columns=["영희", "철수"], index = index)
growth
```

```python
#DataFrame 방법2
index = ["2018", "2019", "2020", "2021"]

Y = pd.Series([143, 150, 157, 160], index = index)
C = pd.Series([165, 172, 175, 180], index = index)

growth = pd.DataFrame({
    "영희":Y,
    "철수":C
})
growth
```

```python
#특정 값만 출력
index = ["2018", "2019", "2020", "2021"]

Y = pd.Series([143, 150, 157, 160], index = index)
C = pd.Series([165, 172, 175, 180], index = index)
S = pd.Series([167, 178, 179, 188], index = index)

growth = pd.DataFrame({
    "영희":Y,
    "철수":C,
    "영철":S
})
growth
```

```python
growth["영희"]
```

```python
#인덱스 값 설정 가능
growth["영희"][2:]
```

```python
#iloc 첫번째 행을 의미한다.
growth.iloc[0]
```

```python
#n번째 행,열만 표현하기
growth.iloc[:2, :3]
```

```python
growth.iloc[-2:]
```
